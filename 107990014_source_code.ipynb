```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics

from sklearn import preprocessing
```

```python
data = pd.read_table('/content/household_power_consumption.zip', 
                     compression='zip', sep=';', 
                     low_memory=False, header=0, index_col=False,
                     parse_dates=[0])
```

```python
df = pd.DataFrame(data, dtype=None)
df.replace('?', np.nan,inplace=True)
df
```

```python
df.fillna(0)
```

```python
#convert values to float format

df["Global_active_power"]   = df["Global_active_power"].astype(float)
df["Global_reactive_power"] = df["Global_reactive_power"].astype(float)
df["Voltage"]               = df["Voltage"].astype(float)
df["Global_intensity"]      = df["Global_intensity"].astype(float)
df["Sub_metering_1"]        = df["Sub_metering_1"].astype(float)
df["Sub_metering_2"]        = df["Sub_metering_2"].astype(float)
df["Sub_metering_3"]        = df["Sub_metering_3"].astype(float)
```

```python
print("Max Values\n")
df.iloc[:,2:6].max()
```

```python
print("Min Values\n")
df.iloc[:,2:6].min()
```

```python
print("Count\n")
df.iloc[:,2:6].count()
```

```python
print("Mean Values\n")
df.iloc[:,2:6].mean()
```

```python
print("Standard Deviation\n")
df.iloc[:,2:6].std()
```

```python
minmax_rescale = preprocessing.MinMaxScaler()
minmax_normalize = minmax_rescale.fit_transform(df.iloc[:,2:6])
print(minmax_normalize)
```

```python
df = pd.DataFrame(minmax_normalize)
df.to_csv('normalized_107990014.csv')
```
